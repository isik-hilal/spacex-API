Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the data
data=pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")
data.head()
```

#### Plotting the graphs

We can plot out the <code>FlightNumber</code> vs. <code>PayloadMass</code>and overlay the outcome of the launch.

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=data, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

for the VAFB-SLC  launchsite there are no  rockets  launched for  heavypayload mass(greater than 10000).

#### Flight Number and Launch Site

```python
sns.catplot(x="LaunchSite", y="FlightNumber", hue="Class", data=data, aspect = 5)
plt.ylabel("Flight Number",fontsize=20)
plt.xlabel("LaunchSite",fontsize=20)
plt.show()
```

```python
# Plot a scatter point chart with x axis to be Pay Load Mass (kg) and y axis to be the launch site, and hue to be the class value
sns.catplot(y="PayloadMass", x="LaunchSite", hue="Class", data=data, aspect = 5)
plt.xlabel("LaunchSite",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

We see that different launch sites have different success rates.

#### Success rate of each Orbit type

```python
Creating the bar plot
```

#### FlightNumber vs Orbit type

```python
sns.catplot(x='FlightNumber',y='Orbit',data=data, colorhue='Class')
plt.xlabel('Flight Number')
plt.ylabel('Orbit Details')
plt.show()
```

#### The launch success vs yearly trend

```python
# A function to Extract years from the date 
year=[]
def Extract_year(date):
    for i in data["Date"]:
        year.append(i.split("-")[0])
    return year
```

```python
Extract_year(1) # 1 is the position
data["Year"]=year
average = data.groupby(by="Year").mean()
average.reset_index(inplace=True)
```

```python
plt.plot(average["Year"],average["Class"])
plt.xlabel("Year")
plt.ylabel("Success/Failure")
plt.show()
```

```python
features = data[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

```python
features_dummy=pd.get_dummies(features, columns=['Orbit','LaunchSite', 'LandingPad', 'Serial'])
features_dummy
```
